Before proceeding to the next section, it is essential to address the missing values. Standard imputation techniques involve using the mean, median, or mode; however, I implemented a more tailored approach. Specifically, the imputation process was customized based on the vehicle make, which, although exhaustive and time-consuming, was intended to enhance model accuracy significantly. Each vehicle make had its own specific mean, median, and mode calculated. 

This data alteration is entirely customized and, while beneficial for accuracy, is not generally recommended due to its time-intensive nature.

# Audi

```python
auto[auto['make']=='audi']
```

Here we have 3 zero values for AUDI in the normalized-lossess column we will solve this by filling it with its median and handle the price at the end to be zero situation

```python
# finding the median of the column normalized-losses with the make to be Audi
audi_nrml_ls_median = auto[auto['make']=='audi']['normalized-losses'].median()
```

```python
audi_nrml_ls_median
```

```python
# audi_normalized_losses = 0 = median

auto.loc[(auto['make']=='audi') & (auto['normalized-losses']==0), 
         'normalized-losses']  = int(audi_nrml_ls_median)


# df[(df['make'] == 'audi') & (df['normalized-losses'] == 0)]
```

```python
auto.dropna(subset=['price'],inplace=True)
```

```python
auto.drop(auto[auto['price']==0].index,axis=0,inplace=True)
```

```python
auto[auto['make']=='audi']
```
